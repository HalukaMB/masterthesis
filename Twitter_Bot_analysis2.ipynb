```python
import pandas as pd
import re
import io
```

```python
data = pd.read_csv('/Users/halukamaier-borst/Documents/tweets.csv', skipinitialspace = True, quotechar = "'")
print(data.head(5))
```

```python
data=data.sort_values(by='id')
print(data.head(5))
```

```python
id=0
counter=0
botlist=[]
for cell in data["id"]:
    if counter>49:
        if id not in botlist:
            botlist.append(id)
        else:
            pass
    if cell != id:
        id=cell
        counter=0
    else:
        counter+=1

print(botlist)
print(len(botlist))
```

```python
frames=[data["text"], data["id"],]
reduced_data=pd.concat(frames, axis=1)
print(reduced_data.head(5))
```

```python
botsample=[]
normalsample=[]
for index, row in reduced_data.iterrows():
    if(row['id'])in botlist:
        botsample.append(row['text'])
    else:
        normalsample.append(row['text'])
print(botsample[0:5])
```

```python
botdf=pd.DataFrame()
botdf["text"]=botsample
botdf["botornot"]=1
normaldf=pd.DataFrame()
normaldf["text"]=normalsample
normaldf["botornot"]=0
normaldf.reset_index(drop=True, inplace=True)
botdf.reset_index(drop=True, inplace=True)
print(normaldf.head(5))
print(botdf.head(5))
```
