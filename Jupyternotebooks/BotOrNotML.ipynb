```python
import sklearn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
traindata = pd.read_csv('/Users/halukamaier-borst/Documents/CardiffUniversity/Master Thesis/botornotREL/botornotsaturdayREL.csv', sep=',',quotechar="'", 
                  engine='python')
```

```python
traindata.describe()
```

```python
traindata=traindata.drop("id",axis=1)
```

```python
traindata=traindata.dropna(how='any')
traindata['botornot'].replace(0, -1,inplace=True)
traindata.describe()
```

```python
Y=traindata.pop("botornot")
X=traindata
traindata.describe()
```

```python
testdata = pd.read_csv('/Users/halukamaier-borst/Documents/CardiffUniversity/Master Thesis/botornotREL/botornotwednesdayREL.csv', sep=',',quotechar="'", 
                  engine='python')
```

```python
testdata=testdata.drop("id",axis=1)
testdata['botornot'].replace(0, -1,inplace=True)
```

```python
Y_test=testdata.pop("botornot")
X_test=testdata
testdata.describe()
```

```python
dtree = DecisionTreeClassifier(criterion='gini', splitter='best', max_features=0.6, max_depth=5, min_samples_split=2, min_samples_leaf=2, min_weight_fraction_leaf=0.0, random_state=None, max_leaf_nodes=None, min_impurity_split=0.2, class_weight=None, presort=False)
```

```python
dtree.fit(X,Y)
```

```python
from sklearn import metrics
metrics.mean_absolute_error(Y_test, dtree.predict(X_test))
```

```python
metrics.accuracy_score(Y_test, dtree.predict(X_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=15, max_features='auto', max_leaf_nodes=None,
            min_impurity_split=1e-07, min_samples_leaf=1,
            min_samples_split=2, min_weight_fraction_leaf=0.0,
            n_estimators=10, n_jobs=1, oob_score=False, random_state=None,
            verbose=0, warm_start=False)
```

```python
rf.fit(X,Y)
```

```python
metrics.mean_absolute_error(Y_test, rf.predict_proba(X_test)[:, 1])
```

```python
from sklearn import svm
clf = svm.SVC()
```

```python
clf.fit(X,Y)
metrics.mean_absolute_error(Y_test, clf.predict(X_test))
```

```python
metrics.accuracy_score(Y_test, clf.predict(X_test))
```
