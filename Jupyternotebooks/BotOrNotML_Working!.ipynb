```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics


def kfold(clr,X,y,folds=10):
    from sklearn.cross_validation import StratifiedKFold
    from sklearn import metrics
    mae_sum=0
    acc_sum=0
    kf = StratifiedKFold(y, folds)
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        clr.fit(X_train, y_train)
        pred_test = clr.predict(X_test)
        print (metrics.mean_absolute_error(y_test,pred_test))
        print (metrics.accuracy_score(y_test,pred_test))
        mae_sum+=metrics.mean_absolute_error(y_test,pred_test)
        acc_sum+=metrics.accuracy_score(y_test,pred_test)
    print ('MAE: ',  mae_sum/folds)
    print ('ACC: ',  acc_sum/folds)



#read the dataset
X = pd.read_csv('/Users/halukamaier-borst/Documents/CardiffUniversity/Master Thesis/botornotREL/botornotmondayREL.csv', sep=',',quotechar="'", 
                  engine='python')

y=X['botornot']

#changes N, and Y to 0, and 1 respectively

del X['botornot']
del X['id']
print(X.head())

#initialize random forests (by defualt it is set to 10 trees)
rf=RandomForestClassifier()

#run algorithm
kfold(rf,np.array(X),y)

#You will get an average AUC around 0.62 as opposed to 0.79 in WEKA
```
